```python
import numpy as np
import math as math
import pandas as pd
# Load the CSV file into a DataFrame
hf=pd.read_csv('C:/Users/ksrad/Downloads/Yogesh Gupta.csv')
hf
FIN=hf.drop(['Authors','Author(s) ID','Volume','Issue','Art. No.','Page start','Page end','Page count','DOI','Link','Document Type','Publication Stage','Open Access','Source','EID'],axis=1)
FIN
FIN.rename(columns = {'Cited by':'C'}, inplace = True)
FIN['h'] = pd.Series([2 for x in range(len(FIN.index))])
#FIN = FIN[:-1]
FIN['C'].replace(np.nan, 0, inplace=True)
FIN
```

```python
def impact(C, h, Year):
    k = 0.001  
    while C >= (h + 1) ** k:
        k += 0.00001  
    return k
FIN['k'] = FIN.apply(lambda row: impact(row['C'], row['h'], row['Year']), axis=1)
FIN['date'] = 2023 - FIN['Year']
FIN['kprime'] = FIN['k'] / FIN['date']
print(FIN)
FIN['kprime'].replace(np.inf, 0, inplace=True)
FIN
```

```python
kp = FIN['kprime'].sum()
Kzlog = np.log(kp)
print(Kzlog)
```
